# Load results

```python
base_config = {
    'results_folder': '../results/results_crossroad/'
}
```

```python
s_dictio = {}

for run in os.listdir(base_config['results_folder']):

    if '.' in run: continue
    if run.split('_')[-1] == 'ae': continue

    r_path = os.path.join(base_config['results_folder'], run)
    print(r_path)

    res_d = {'ADE': [], 'FDE': [], 'des': [], 'preds': [], 'col': [], 'config': []} #, 'col_per': []}

    list_r = [x for x in os.listdir(r_path) if 'npy' in x and 'preds' in x and 'cluster' not in x]

    print(len(list_r))
    if len(list_r) == 0: continue

    for x in list_r:

        preds = np.load(os.path.join(r_path, x))
        config_path = os.path.join(r_path, x.replace('preds', 'config'))
        config_path = config_path.replace('npy', 'yaml')
        config = load_config_file(config_path, 0)
        config.update(base_config)
        config['data_path'] = os.path.join('..', config['data_path'])
        config['semantic_scene_folder'] = os.path.join(config['data_path'], 'segmentations')
        data = LoadData(config, verbose = 0)
        
        preds = np.cumsum(preds, axis = -2)
        data.Y_test = np.cumsum(data.Y_test, -2)
                
        # global positions
        
        for i in range(preds.shape[1]):
            for j in range(preds.shape[2]):
                preds[:, i, j] += data.X_test['obs_traj'][:, -1]
        
        for i in range(data.Y_test.shape[1]):
            data.Y_test[:, i] += data.X_test['obs_traj'][:, -1]
                
        res_d['preds'].append(preds)
        res_d['config'].append(config)
        
        # collisions
                
        scene = get_scene(config["semantic_scene_folder"], config['i_test'])
        scene = np.argmax(scene, axis = -1)
                
        pred_pix = tf.convert_to_tensor(preds.astype('int32'))
        ss_map = tf.convert_to_tensor(scene)
        
        pred_pix = tf.reshape(pred_pix, [-1, 2])
        indices = tf.concat([pred_pix[:, 1:2], pred_pix[:, 0:1]], axis = -1)

        some = tf.gather_nd(
            ss_map, indices,
        )
        
        some = tf.reshape(some, preds.shape[:-1])
        some = 1 - np.array(some).astype('float32')
                
        fact_d = 1.0*some.shape[-1] * some.shape[-2]
        some = np.sum(some, axis = -1)
        some = np.sum(some, axis = -1)
        
        res_d['col'].append(some)
        
        # variance
        
        lastp = preds[:, :, -1, :]
        difs = np.zeros_like(lastp)
        means = np.mean(lastp, axis = 1)
        for i in range(difs.shape[1]):
            difs[:, i, :] = lastp[:, i, :] - means
        norms = np.linalg.norm(difs, axis = -1)
        des = np.std(norms, axis = -1)
        res_d['des'].append(des)

        # ade and fde
        
        _, a, b = get_metrics(preds, data.Y_test, rela = False)
        
        res_d["ADE"].append(a)
        res_d["FDE"].append(b)
        
    if len(res_d) > 0:
        if not run in s_dictio:
            s_dictio[run] = {}

        s_dictio[run] = res_d
```

# Results

```python
for run in s_dictio:
    
    print("************************", run)
    
    res_d = s_dictio[run]

    x_m = {'ADE': [], 'FDE': [], 'des': [], 'col': []} # , 'col_per': []}
    
    for x in res_d:
        
        if x in ['preds', 'preds_pix', 'config']:
            
            continue
            
        for lx in res_d[x]:
            x_m[x].append(np.mean(lx))

    for x in x_m:
        
        line_p = "{0:.4}\t{1:.2}\t{2:.3}\t{3:.3}".format(np.mean(x_m[x]), np.std(x_m[x]), np.min(x_m[x]), np.max(x_m[x]))
        
        print(line_p, end = '\n')
        #print(line_p, end = '\t')
        
        line_p = "{0:.3}"
        for ii in range(1, len(x_m[x])): line_p += ", {" + str(ii) + ':.3}'
        line_p = line_p.format(*x_m[x])
        # print(line_p)
        # print()

    print()
        
```

# Trajectory visualization

```python
def plot_trajs(config, X, Y, ind, scene = str(config['i_test'])):

  dir_im = os.path.join(config["semantic_scene_folder"], scene + '.npy')
  frame = np.load(dir_im).astype('float64')
  frame_trajs = np.zeros_like(frame).astype('uint8')

  trajs_ok = X[ind]
  traj_gt = Y[ind]
  trajs_ = data.X_test['obs_traj'][ind]
       
  for traj in trajs_ok:
    aux_frame = np.zeros_like(frame_trajs)
    for ii in range(1, len(traj)):
      xf, yf = traj[ii]
      xi, yi = traj[ii-1]
      # image = cv2.circle(image, center_coordinates, radius, color, thickness)
      start_point = (int(xi), int(yi))
      end_point = (int(xf), int(yf))
      cv2.line(aux_frame, start_point, end_point, 20, 2)
    frame_trajs += aux_frame
      
  frame_trajs = cv2.blur(frame_trajs, (10, 10))
  frame_trajs = frame_trajs.astype('float64')/np.max(frame_trajs)

  for (x, y) in trajs_:
    cv2.circle(frame_trajs, (int(x), int(y)), 4, 0.5, -1)

  for (x, y) in traj_gt:
    cv2.circle(frame_trajs, (int(x), int(y)), 4, 1, -1)

  frame = frame.astype('float64')/np.max(frame)
    
  frame = (1-(frame_trajs>0))*frame
    
  frame = cv2.addWeighted(frame, 1, frame_trajs, 1, 0)
  frame = 255*frame.astype('float64')/np.max(frame)
      
  print(config['use_features'])
        
  if config['use_features']:
    
    x, y = trajs_[-1]
    x, y = int(x), int(y)
    gs = config['nLGrid']
    
    top_left = (x-gs//2, y-gs//2)
    bottom_right = (x+gs//2, y+gs//2)
    
    image = cv2.rectangle(frame, top_left, bottom_right, 80, 2)
    
  return frame


crit = 'des'
res_s = 'vae_0_1_1'
ind = 0

# res_s = 'sdvae_1_1'
# ind = 3

# res_s = 'sdvae_0_1'
# ind = 4 # 1


break_f = False
# break_f = True

# int_2_1_1 = [499, 1664] # bad cases
# int_2_1_1 = [138, 1283, 1317] # ok cases
int_2_1_1 = [3508,  215,  104,  102,  499,  692, 3510, 3572,  217, 4044]

imas = []

# for s_ind in [1362, 1904, 534, 4171, 1317, 1283, 2441, 138]:
# for s_ind in range(0,80):
# for s_ind in np.argsort(s_dictio[res_s][crit][ind])[-40:-20]:
# for s_ind in [np.argsort(s_dictio[res_s][crit][ind])[0], np.argsort(s_dictio[res_s][crit][ind])[-1]]:
for s_ind in int_2_1_1:

    if break_f:
        s_ind = np.argsort(s_dictio[res_s][crit][ind])[0]
        # s_ind = 1110
        print(s_ind)

    print(s_ind)
    print(s_dictio[res_s][crit][ind][s_ind])
    fr = plot_trajs(s_dictio[res_s]['config'][ind], s_dictio[res_s]['preds'][ind], data.Y_test, s_ind)

    imas.append(fr)
    rcParams['figure.figsize'] = 8, 10
    plt.axis('off')
    plt.imshow(fr)
    plt.show()

    if break_f: break


imas = np.hstack(imas)

rcParams['figure.figsize'] = 15, 20
plt.axis('off')
plt.imshow(imas)
plt.show()
```

# Compare SDVAE encoder and encoder from scratch

```python
dir_res = os.path.join('..', *config['r_path'].split('/')[:-1], 'sdvae' + '_0_1_0')
list_f = [x for x in os.listdir(dir_res) if 'config' in x]

assert len(list_f) > 0

list_f.sort()

ae_config = load_config_file(os.path.join(dir_res, list_f[0]), 0)

ae_config['ae_epochs'] = 3
ae_config['ae_lr'] = 0.001 
ae_config['ae_augmentation'] = False

h_list = []
imas = []

for freeze in range(2):

    tf.keras.backend.clear_session()
    autoencoder = Autoencoder(ae_config)

    if freeze:

        enc_path = os.path.join(dir_res, '_'.join(['ss_enc'] + list_f[0].split('_')[:3]) + '.h5')
        autoencoder.planar_encoder.load_weights(enc_path)

        for layer in autoencoder.conv_enc:

            layer.trainable = False
            
    data = LoadData(config)
    ae_file = os.path.join(dir_res + '_ae')
    
    vali_data = PatchesDataGenerator(data.X_test, ae_config)
        
    h, _, _ = autoencoder.train(ae_config, data)
    h_list.append(h.history)
    
    print("Freeze:", freeze)
    
    ids = [1, 50, 100, 150, 200, 250]

    plt.rcParams['figure.figsize'] = [5, 5]

    batch = vali_data.data_generation(ids)
    reco = autoencoder.ae.predict(batch)
    
    i = 0
    
    for (ima, ima_2) in zip(reco, batch):

        ima = np.argmax(ima, axis=-1)
        ima_2 = np.argmax(ima_2, axis=-1)

        if freeze == 0:
            
            imas.append([ima_2, ima])
            
        else:
            
            imas[i].append(ima)
        
        i += 1

plt.rcParams['figure.figsize'] = [6.5, 5]
plt.title('AE Training')
    
plt.xlabel('Epoch')
plt.ylabel('Cross-Entropy')
    
for (i, h) in enumerate(h_list):

    print(i)
    
    if i == 0:
        name = 'From scratch.'
    else:
        name = 'SDVAE.'
    
    n = len(h['loss'])
    plt.plot(range(1, n+1), h['loss'], label = name + ' Train')
    plt.plot(range(1, n+1), h['val_loss'], label = name + ' Test')
    
plt.legend()
name_f = os.path.join('..', *config['r_path'].split('/')[:-1], 'ae_training.pdf')
plt.savefig(name_f, format = 'pdf')
plt.show()
print(name_f)

for (i, ims) in enumerate(imas):
    
    if i%8 != 0: continue
    
    # plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, hspace = 0, wspace = 0)
    # plt.margins(0,0)
    
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3)    
    fig.set_size_inches(12.0, 4)
    
    ax1.set_title('Original patch.')    
    ax2.set_title('AE reconstruction.')
    ax3.set_title('Reconstruction from SDVAE features.')

    ax1.axis('off')
    ax2.axis('off')
    ax3.axis('off')
    
    ims[0] = np.clip(ims[0], 0.25, 1)
    ims[1] = np.clip(ims[1], 0.25, 1)
    ims[2] = np.clip(ims[2], 0.25, 1)
    
    ax1.imshow(ims[0], cmap = 'Blues', vmin = 0, vmax = 1)
    ax2.imshow(ims[1], cmap = 'Blues', vmin = 0, vmax = 1)
    ax3.imshow(ims[2], cmap = 'Blues', vmin = 0, vmax = 1)
    
    name_f = os.path.join('..', *config['r_path'].split('/')[:-1], 'ae_r{0}.pdf'.format(i//8))
    plt.savefig(name_f, format = 'pdf', bbox_inches = 'tight')

    fig.show()
  
```
