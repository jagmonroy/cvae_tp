# Qualitative results: best, middle and worst cases

```python
# if there are many predictions, 'plot_dist' can be used modifying a 'plot_index' parameter (set dist to true).

def get_limits(preds, ax):
                
    preds = np.array(preds)
    return np.array([np.min(preds[..., :, ax]), np.max(preds[..., :, ax])])


def plot_dist(obs, gt, preds, n_grid = 1000):

    lx = get_limits(preds, 0)
    ly = get_limits(preds, 1)

    lxy = [np.min([lx[0], ly[0]]), np.max([lx[1], ly[1]])]

    grid = np.zeros((n_grid, n_grid))

    for pred in preds:

        pred[:, 0] = (pred[:, 0] - lx[0])/(lx[1] - lx[0]) * (n_grid-1)
        pred[:, 1] = (pred[:, 1] - ly[0])/(ly[1] - ly[0]) * (n_grid-1)
        pred[:, 1] = (n_grid-1) - pred[:, 1]

        pred = pred.astype('int')
        grid_aux = np.zeros_like(grid)
        pred = [tuple(x) for x in pred]

        for i in range(1, len(pred)):
            cv2.line(grid_aux, pred[i], pred[i-1], 1, 75)

        grid += grid_aux

    backg = grid<1

    grid[backg] = np.nan

    plt.axis('equal')
    plt.imshow(grid, aspect = 'auto', extent = [*lx, *ly],
               # vmin = np.min(grid), vmax = np.max(grid),
               cmap = 'Blues')

    plt.plot(obs[:, 0], obs[:, 1], c = 'b', marker = 'o', label = 'Obs')
    plt.plot(gt[:, 0], gt[:, 1], c = 'r', marker = 'o', label = 'GT')

    plt.legend()
    plt.show()
```
